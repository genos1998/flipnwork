### Question 2
### IMDB top 100

```python
page=requests.get('https://www.imdb.com/chart/top')
#page
```

```python
soup=BeautifulSoup(page.content)
```

```python
titles=soup.find_all('td',class_="titleColumn")
tit=[]
for i in titles:
    for j in i.find_all('a'):
        tit.append(j.text.replace('\n',' '))
#len(tit[:100])
```

```python
rate=soup.find_all('td',class_="ratingColumn imdbRating")
rating=[]
for i in rate:
    rating.append(i.text.strip())
#rating[:100]
```

```python
yearly=soup.find_all('td',class_="titleColumn")
year=[]
for i in yearly:
    for j in i.find_all('span',class_='secondaryInfo'):
        year.append(j.text.strip().replace('(',''))
for i in range(len(year)):
    year[i]=year[i].replace(')','')
#year[:100]
```

```python
imdb=pd.DataFrame({})
imdb['Name']=tit[:100]
imdb['Rating']=rating[:100]
imdb['Year']=year[:100]
imdb
```

### Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
soup=BeautifulSoup(page.content)
```

```python
titles=soup.find_all('div',class_="heading_4_5 profile")
job_titles=[]
for i in titles:
    job_titles.append(i.text.replace('\n',''))
job_titles[1:]
```

```python
companies=soup.find_all('a',class_='link_display_like_text')
company=[]
for i in companies:
    company.append(i.text.strip())
company[1:]
```

```python
ctc=soup.find_all('div',class_='item_body')
CTC=[]
for i in ctc:
    CTC.append(i.text.strip())
CTC[4::3]
```

```python
apply=soup.find_all('div',class_='item_body')
date=[]
for i in apply:
    date.append(i.text.strip())
date[5::3]
```

```python
internshala=pd.DataFrame({})
internshala['Job Title']=job_titles[1:]
internshala['Company']=company[1:]
internshala['CTC']=CTC[4::3]
internshala['apply date']=date[5::3]
internshala
```
