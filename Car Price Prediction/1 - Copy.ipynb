```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
import ipyparallel as ipp
```

```python
link=pd.read_csv('cardekho_link.csv')
link
```

```python
urls=list(link['url'].values[1000:2000])
year=[]
fuel=[]
kms=[]
owners=[]
transmission=[]
price=[]
mileage=[]
engine=[]
power=[]
seats=[]
gear=[]
drive=[]
steering=[]
torque=[]
values_cylinder=[]
no_cylinder=[]
turbo=[]
super_charger=[]
i=0
ui=0
driver=webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
for url in urls:
    driver.get(url)
    time.sleep(3)
    driver.execute_script("window.scrollBy(0,1000)")
    time.sleep(2)
    driver.execute_script("window.scrollBy(0,500)")
    time.sleep(2)
    try:
        if driver.find_element_by_xpath('//div[@class="hover whatsappShare margt25 shareButton"]').text=='View All Specifications':
            button=driver.find_element_by_xpath('//div[@class="hover whatsappShare margt25 shareButton"]')
            button.click()
            ui=0
            time.sleep(2) 
        else:
            raise NoSuchElementException
    except:
        try:
            view=driver.find_element_by_xpath('//div[@class="brdrBtm"]/a')
            view.click()
            ui=1
            time.sleep(2) 
        except:
            try:
                view=driver.find_element_by_xpath('//a[@class=" viewMorebtn viewMorebtnArrow"]')
                view.click()
                ui=1
                time.sleep(2)
            except NoSuchElementException:
                i+=1
                continue
        #--year--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[2]/div/div[2]')      
        year.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[1]/div/div')      
            year.append(ear.text) 
        except:
            year.append('-')
        #--fuel--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[3]/div/div[2]')      
        fuel.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[5]/div/div') 
            fuel.append(ear.text) 
        except:
            fuel.append('-')
    #--kms--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[4]/div/div[2]')      
        kms.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[3]/div/div') 
            kms.append(ear.text) 
        except:
            kms.append('-')
    #--owners--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[6]/div/div[2]')      
        owners.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[4]/div/div') 
            owners.append(ear.text) 
        except:
            owners.append('-')
    #--transmission--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[8]/div/div[2]')      
        transmission.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[3]/div/div/div/div[1]/ul/li[6]/div/div') 
            transmission.append(ear.text) 
        except:
            transmission.append('-')
    #--price--
    try:
        ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div/div/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div[1]/span[2]')      
        price.append(ear.text)
    except NoSuchElementException:
        try:
            ear = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/main/div[2]/div[1]/div/div[2]/div[2]/div/div/div/div[1]/span/span') 
            price.append(ear.text) 
        except:
            price.append('-')
    if ui==0:
        items=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-6 gsc_col-md-4"]')
        first=0
        powerr=0
        seatss=0
        steeringg=0
        torquee=0
        ncy=0
        vcy=0
        turboo=0
        superr=0
        gearr=0
        drivee=0
        mileagee=0
        enginee=0
        for j in range(len(items)):
            if len(items[j].text.split())>1:
                #print(items[j].text.split())
                if items[j].text.split()[1]=='Power' and first==0:
                    first=1
                    continue
                elif len(items[j].text.split())>2 and items[j].text.split()[1]=='Power' and first==1:
                    power.append(items[j].text.split()[2])
                    powerr=1
                elif items[j].text.split()[0]=='Seats':
                    seats.append(items[j].text.split()[1])
                    seatss=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Steering':
                    steering.append(items[j].text.split()[2])
                    steeringg=1
                elif len(items[j].text.split())>2 and items[j].text.split()[1]=='Torque':
                    torque.append(items[j].text.split()[2])
                    torquee=1
                elif len(items[j].text.split())>3 and items[j].text.split()[0]=='No' and items[j].text.split()[2]=='Cylinder':
                    no_cylinder.append(items[j].text.split()[3])
                    ncy=1
                elif len(items[j].text.split())>3 and items[j].text.split()[0]=='Values':
                    values_cylinder.append(items[j].text.split()[3])
                    vcy=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Turbo':
                    turbo.append(items[j].text.split()[2])
                    turboo=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Super':
                    super_charger.append(items[j].text.split()[2])
                    superr=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Gear':
                    gear.append(items[j].text.split()[2])
                    gearr=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Drive':
                    drive.append(items[j].text.split()[2])
                    drivee=1
                elif items[j].text.split()[0]=='Mileage':
                    mileage.append(items[j].text.split()[1])
                    engine.append(items[j+1].text.split()[1])
                    mileagee=1
                    enginee=1

        if powerr==0:
            power.append('-')
        if seatss==0:
            seats.append('-')
        if steeringg==0:
            steering.append('-')
        if torquee==0:
            torque.append('-')
        if ncy==0:
            no_cylinder.append('-')
        if vcy==0:
            values_cylinder.append('-')
        if turboo==0:
            turbo.append('-')
        if superr==0:
            super_charger.append('-')
        if gearr==0:
            gear.append('-')
        if drivee==0:
            drive.append('-')
        if enginee==0:
            engine.append('-')
        if mileagee==0:
            mileage.append('-')
    elif ui==1:
        items=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-6"]')
        first=0
        powerr=0
        seatss=0
        steeringg=0
        torquee=0
        ncy=0
        vcy=0
        turboo=0
        superr=0
        gearr=0
        drivee=0
        mileagee=0
        enginee=0
        for j in range(len(items)):
            if len(items[j].text.split())>1:
                if items[j].text.split()[0]=='Seats':
                    seats.append(items[j].text.split()[1])
                    seatss=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Steering':
                    steering.append(items[j].text.split()[2])
                    steeringg=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Max' and items[j].text.split()[1]=='Torque':
                    torque.append(''.join(items[j].text.split()[2:]))
                    power.append(''.join(items[j-1].text.split()[2:]))
                    powerr=1
                    torquee=1
                elif len(items[j].text.split())>3 and items[j].text.split()[0]=='No' and items[j].text.split()[2]=='Cylinder':
                    no_cylinder.append(items[j].text.split()[3])
                    ncy=1
                elif len(items[j].text.split())>3 and items[j].text.split()[0]=='Valves':
                    values_cylinder.append(items[j].text.split()[3])
                    vcy=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Turbo':
                    turbo.append(items[j].text.split()[2])
                    turboo=1
                elif items[j].text.split()[0]=='superCharger':
                    super_charger.append(items[j].text.split()[1])
                    superr=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Gear':
                    gear.append(items[j].text.split()[2])
                    gearr=1
                elif len(items[j].text.split())>2 and items[j].text.split()[0]=='Drive':
                    drive.append(items[j].text.split()[2])
                    drivee=1
                elif items[j].text.split()[0]=='Mileage':
                    mileage.append(items[j].text.split()[1])
                    engine.append(items[j+1].text.split()[1])
                    mileagee=1
                    enginee=1 
        if powerr==0:
            power.append('-')
        if seatss==0:
            seats.append('-')
        if steeringg==0:
            steering.append('-')
        if torquee==0:
            torque.append('-')
        if ncy==0:
            no_cylinder.append('-')
        if vcy==0:
            values_cylinder.append('-')
        if turboo==0:
            turbo.append('-')
        if superr==0:
            super_charger.append('-')
        if gearr==0:
            gear.append('-')
        if drivee==0:
            drive.append('-')
        if enginee==0:
            engine.append('-')
        if mileagee==0:
            mileage.append('-')
    i+=1
    print(i,end=', ')
```

```python
cardekho=pd.DataFrame({})
cardekho['year']=year
cardekho['fuel']=fuel
cardekho['kms']=kms
cardekho['owners']=owners
cardekho['transmission']=transmission
cardekho['mileage']=mileage
cardekho['engine']=engine
cardekho['power']=power
cardekho['seats']=seats
cardekho['gear']=gear
cardekho['drive']=drive
cardekho['steering']=steering
cardekho['torque']=torque
cardekho['values_cylinder']=values_cylinder
cardekho['no_cylinder']=no_cylinder
cardekho['turbo']=turbo
cardekho['super_charger']=super_charger
cardekho['price']=price
cardekho
```

```python
data=pd.read_csv('cardekho.csv')
data=data.drop('Unnamed: 0',axis=1)
data
```

```python
vertical_concat = pd.concat([data,cardekho ], axis=0)
vertical_concat
```
