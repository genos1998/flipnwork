```python
data['power_rpm']=0
data['torque_rpm']=0
for i in range(len(data.kms.values)):
    data['kms'].iloc[i]=data.kms.values[i].split()[0]
    if ''.join(data.kms.values[i].split(',')).isdigit():
        data['kms'].iloc[i]=int(''.join(data.kms.values[i].split(',')))
for i in range(len(data.owners.values)):
    if data.owners.values[i].split()[1]=='&':
        data['owners'].iloc[i]=data.owners.values[i].split()[0]+'+'
    else:
        data['owners'].iloc[i]=data.owners.values[i].split()[0]
for i in range(len(data.drive.values)):
    if data.drive.values[i]=='2' or data.drive.values[i]=='2WD' or data.drive.values[i]=='2wd' or data.drive.values[i]=='Two' or data.drive.values[i]=='4X2' or data.drive.values[i]=='Front':
        data.iloc[i]='FWD'
    elif data.drive.values[i]=='4WD' or data.drive.values[i]=='4X4' or data.drive.values[i]=='All' or data.drive.values[i]=='All-wheel':
        data.iloc[i]='AWD'
    elif data.drive.values[i]=='Rear':
        data.iloc[i]='RWD'
for i in range(len(data.power.values)):
    try:
        if len(data.power.values[i].split('@'))==1 and len(data.power.values[i].split('PSat'))==1:
            data['power'].iloc[i]=0
        else:
            data['power_rpm'].iloc[i]=data.power.values[i].split('@')[1]
            data['power'].iloc[i]=data.power.values[i].split('@')[0]
    except:
        data['power'].iloc[i]=0
        data['power_rpm'].iloc[i]=0
for i in range(len(data.power.values)):
    try:
        if data.power.values[i][-3:].lower()=='bhp':
            data['power'].iloc[i]=float(data.power.values[i][:-3])
        elif data.power.values[i][-2:]=='PS':
            data['power'].iloc[i]=float(data.power.values[i][:-2])
    except:
        pass
for i in range(len(data.power_rpm.values)):
    try:
        if data.power_rpm.values[i][-3:].lower()=='rpm' or data.power_rpm.values[i][-3:].lower()=='prm':
            data['power_rpm'].iloc[i]=float(data.power_rpm.values[i][:-3])
        else:
            if data.power_rpm.values[i]=='FWD' or data.power_rpm.values[i]=='AWD' or data.power_rpm.values[i]=='RWD':
                data['power_rpm'].iloc[i]=0
            else:
                if data['power_rpm'].iloc[i].split('(')[0].isdigit():
                    data['power_rpm'].iloc[i]=data['power_rpm'].values[i].split('(')[0]
                else:
                    if ''.join(data['power_rpm'].iloc[i].split(',')[0]).isdigit():
                        data['power_rpm'].iloc[i]=''.join(data['power_rpm'].iloc[i].split(',')[0])
                    else:
                        data['power_rpm'].iloc[i]=data.power_rpm.values[i].split('-')[0]
                        
    except:
        if data.power_rpm.values[i]=='FWD' or data.power_rpm.values[i]=='AWD':
            data['power_rpm'].iloc[i]=0
for i in range(len(data.torque.values)):
    try:
        if len(data.torque.values[i].split('@'))==1:# and len(data.torque.values[i].split('PSat'))==1:
            data['torque'].iloc[i]=0
        else:
            data['torque_rpm'].iloc[i]=data.torque.values[i].split('@')[1]
            data['torque'].iloc[i]=data.torque.values[i].split('@')[0]
    except:
        data['torque'].iloc[i]=0
        data['torque_rpm'].iloc[i]=0
for i in range(len(data.torque.values)):
    try:
        if data.torque.values[i][-2:].lower()=='nm':
            data['torque'].iloc[i]=data.torque.values[i][:-2]
        elif data.torque.values[i][-3:].lower()=='kgm':
            data['torque'].iloc[i]=data.torque.values[i][:-3]
        elif data.torque.values[i][-4:].lower()=='kg-m':
            data['torque'].iloc[i]=data.torque.values[i][:-4]
    except:
        pass
for i in range(len(data.torque_rpm.values)):
    try:
        if data.torque_rpm.values[i][-3:]=='rpm':
            data['torque_rpm'].iloc[i]=data.torque_rpm.values[i][:-3]
        elif data.torque_rpm.values[i]=='FWD' or data.torque_rpm.values[i]=='AWD' or data.torque_rpm.values[i]=='RWD':
            data['torque_rpm'].iloc[i]=0 
        elif data.torque_rpm.values[i][-4:]=='(kgm':
            data['torque_rpm'].iloc[i]=data.torque_rpm.values[i][:-4]
        if ''.join(data['torque_rpm'].iloc[i].split(',')[0]).isdigit():
            data['torque_rpm'].iloc[i]=''.join(data['torque_rpm'].iloc[i].split(',')[0])
        if len(data.torque_rpm.values[i].split('-'))>1:
            if data.torque_rpm.values[i].split('-')[1][-3:]=='(NM':
                data['torque_rpm'].iloc[i]=data.torque_rpm.values[i].split('-')[1][:-3]
            else:
                data['torque_rpm'].iloc[i]=data.torque_rpm.values[i].split('-')[1]
        if ''.join(data['torque_rpm'].iloc[i].split(',')[0]).isdigit():
            data['torque_rpm'].iloc[i]=''.join(data['torque_rpm'].iloc[i].split(',')[0])
    except:
        pass
for i in range(len(data.price.values)):
    if data.price.values[i][-5:]=='Lakh*':
        data['price'].iloc[i]=float(data.price.values[i].split()[0])*100000
    elif data.price.values[i]=='FWD' or data.price.values[i]=='AWD' or data.price.values[i]=='RWD':
        data['price'].iloc[i]=0  
    elif data.price.values[i][-1]=='*':
        data['price'].iloc[i]=data['price'].iloc[i][:-1]
    elif ''.join(data.price.values[i].split(',')).isdigit():
        data['price'].iloc[i]=float(''.join(data.price.values[i].split(',')))
dropp=[]
for i in range(len(data['year'].values)):
    try:
        data['year'].iloc[i]=int(data['year'].values[i])
    except:
        if data.year.values[i]=='FWD' or data.year.values[i]=='AWD' or data.year.values[i]=='RWD':
            dropp.append(i)
        else:
            try:
                data['year'].iloc[i]=int(data['year'].values[i].split()[1])
            except:
                print(data['year'].values[i])
data=data.drop(data.index[dropp])
data=data[['year', 'fuel', 'kms', 'owners', 'transmission', 'mileage', 'engine','power','power_rpm', 'seats', 'gear', 'drive', 'steering', 'torque','torque_rpm','values_cylinder', 'no_cylinder', 'turbo', 'super_charger', 'price']]
data=data.reset_index(drop=True)
for i in range(len(data.gear.values)):
    if len(data.gear.values[i].split('-'))>1:
        data['gear'].iloc[i]=data.gear.values[i].split('-')[0]
    elif data.gear.values[i]=='CVT' or data.gear.values[i]=='AGS' or data.gear.values[i]=='iMT':
        data['gear'].iloc[i]=0
    elif data.gear.values[i][-1]=='G':
        data['gear'].iloc[i]=data.gear.values[i][:-1]
data=data.drop(['super_charger'],axis=1)

data
```
