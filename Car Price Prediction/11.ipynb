```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
import ipyparallel as ipp
```

```python
link=pd.read_csv('carwale_links.csv')
link
```

```python
urls=list(link['url'].values[2000:3000])
year=[]
fuel=[]
kms=[]
owners=[]
transmission=[]
price=[]
mileage=[]
engine=[]
power=[]
seats=[]
gear=[]
drive=[]
steering=[]
torque=[]
values_cylinder=[]
no_cylinder=[]
turbo=[]
super_charger=[]
i=0
driver=webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
for url in urls:
    driver.get(url)
    time.sleep(2)
    try:
        while driver.find_element_by_xpath('//li[@data-tabs="specifications"]').text!='Specifications':
            driver.execute_script("window.scrollBy(0,100)")
    except NoSuchElementException:
        i+=1
        continue
    powerr=0
    seatss=0
    steeringg=0
    torquee=0
    turboo=0
    superr=0
    gearr=0
    drivee=0
    mileagee=0
    enginee=0
    pricee=0
    yearr=0
    kmss=0
    ownerr=0
    fuell=0
    
    items=driver.find_element_by_xpath('//div[@class="overview-list padding-bottom10"]')
    listy=items.text.split('\n')
    for j in range(len(listy)):
        if listy[j]=='Price':
            price.append(listy[j+1])
            pricee=1
        elif listy[j]=='Year':
            year.append(listy[j+1])
            yearr=1
        elif listy[j]=='Kilometer':
            kms.append(listy[j+1])
            kmss=1
        elif listy[j]=='Fuel Type':
            fuel.append(listy[j+1])
            fuell=1
        elif listy[j]=='No. of Owner(s)':
            owners.append(listy[j+1])
            ownerr=1
            
    button=driver.find_element_by_xpath('//li[@data-tabs="specifications"]')
    button.click()
    
    items=driver.find_element_by_xpath('//div[@class="cw-tabs-data uc-data hide overviewConatiner"]')
    listy=items.text.split('\n')
    
    for j in range(len(listy)):
        if listy[j]=='Engine':
            value=listy[j+1].split(',')
            values_cylinder.append(value)
            no_cylinder.append(value)
            engine.append(value)
            enginee=1
        elif listy[j]=='Transmission':
            value=listy[j+1].split('-')
            transmission.append(value)
            gear.append(value)
            gearr=1
        elif listy[j]=='Max Power (bhp@rpm)':
            power.append(listy[j+1])
            powerr=1
        elif listy[j]=='Max Torque (Nm@rpm)':
            torque.append(listy[j+1])
            torquee=1
        elif listy[j]=='Mileage (ARAI)':
            mileage.append(listy[j+1])
            mileagee=1
        elif listy[j]=='Drivetrain':
            drive.append(listy[j+1])
            drivee=1
        elif listy[j]=='Turbocharger/Supercharger':
            turbo.append(listy[j+1])
            super_charger.append(listy[j+1])
            turboo=1
        elif listy[j]=='Seating Capacity':
            seats.append(listy[j+1].split()[0])
            seatss=1
        elif listy[j]=='Steering Type':
            steering.append(listy[j+1])
            steeringg=1
    if powerr==0:
        power.append('-')
    if seatss==0:
        seats.append('-')
    if steeringg==0:
        steering.append('-')
    if torquee==0:
        torque.append('-')
    if turboo==0:
        turbo.append('-')
        super_charger.append('-')
    if gearr==0:
        gear.append('-')
        transmission.append('-')
    if drivee==0:
        drive.append('-')
    if enginee==0:
        engine.append('-')
        values_cylinder.append('-')
        no_cylinder.append('-')
    if mileagee==0:
        mileage.append('-')
    if pricee==0:
        price.append('-')
    if yearr==0:
        year.append('-')
    if kmss==0:
        kms.append('-')
    if ownerr==0:
        owner.append('-')
    if fuell==0:
        fuel.append('-')
    
    i+=1
    print(i,end=', ')
```

```python
carwale=pd.DataFrame({})
carwale['year']=year
carwale['fuel']=fuel
carwale['kms']=kms
carwale['owners']=owners
carwale['transmission']=transmission
carwale['mileage']=mileage
carwale['engine']=engine
carwale['power']=power
carwale['power_rpm']=0
carwale['seats']=seats
carwale['gear']=gear
carwale['drive']=drive
carwale['steering']=steering
carwale['torque']=torque
carwale['torque_rpm']=0
carwale['values_cylinder']=values_cylinder
carwale['no_cylinder']=no_cylinder
carwale['turbo']=turbo
carwale['super_charger']=super_charger
carwale['price']=price
carwale
```

```python
for i in range(len(carwale.transmission.values)):
    if carwale.transmission.values[i][0].split()[0]!='AMT':
        carwale['transmission'].iloc[i]=carwale.transmission.values[i][0].split()[0]
    else:
        carwale['transmission'].iloc[i]='Automatic'
for i in range(len(carwale.engine.values)):
    if 'cc' in carwale.engine.values[i][0].split()[:2]:
        carwale['engine'].iloc[i]=int(carwale.engine.values[i][0].split()[0])
    else:
        carwale['engine'].iloc[i]=0
for i in range(len(carwale.gear.values)):
    if carwale.gear.values[i]!='-' and len(carwale.gear.values[i])>1:
        if carwale.gear.values[i][1].split()[0].isdigit():
            carwale['gear'].iloc[i]=int(carwale.gear.values[i][1].split()[0])
        elif carwale.gear.values[i][1].split()[0]=='CVT':
            carwale['gear'].iloc[i]=0
for i in range(len(carwale.steering.values)):
    carwale['steering'].iloc[i]=carwale.steering.values[i].split()[0]
for i in range(len(carwale.no_cylinder.values)):
    if len(carwale.no_cylinder.values[i])>1:
        if carwale.no_cylinder.values[i][1].split()[1]=='Cylinders':
            carwale['no_cylinder'].iloc[i]=carwale.no_cylinder.values[i][1].split()[0]
        else:
            carwale['no_cylinder'].iloc[i]=0
for i in range(len(carwale)):
    updated=0
    if len(carwale.values_cylinder.values[i])>1:
        for j in carwale.values_cylinder.values[i][1:]:
            if len(j.split())>1 and j.split()[1]=='Valves/Cylinder':
                updated=1
                carwale['values_cylinder'].iloc[i]=j.split()[0]
                break
        if updated==0:
            carwale['values_cylinder'].iloc[i]=0
for i in range(len(carwale.super_charger.values)):
    if carwale.super_charger.values[i]=='Yes':
        carwale['super_charger'].iloc[i]=1
    else:
        carwale['super_charger'].iloc[i]=0

for i in range(len(carwale.year.values)):
    carwale['year'].iloc[i]=carwale.year.values[i].split()[1]
for i in range(len(carwale.kms.values)):
    carwale['kms'].iloc[i]=carwale.kms.values[i].split()[0]
for i in range(len(carwale.mileage.values)):
    carwale['mileage'].iloc[i]=carwale.mileage.values[i].split()[0]
for i in range(len(carwale.power.values)):
    if carwale.power.values[i]!='-':
        if carwale.power.values[i].split('@')[1]!='':
            carwale['power_rpm'].iloc[i]=carwale.power.values[i].split('@')[1].split()[0]
            carwale['power'].iloc[i]=carwale.power.values[i].split('@')[0].split()[0]
        elif carwale.power.values[i].split('@')[1]=='':
            carwale['power_rpm'].iloc[i]=0
            carwale['power'].iloc[i]=carwale.power.values[i].split('@')[0].split()[0]
for i in range(len(carwale.torque.values)):
    if carwale.torque.values[i]!='-':
        if carwale.torque.values[i].split('@')[1]!='':
            carwale['torque_rpm'].iloc[i]=carwale.torque.values[i].split('@')[1].split()[0]
            carwale['torque'].iloc[i]=carwale.torque.values[i].split('@')[0].split()[0]
        elif carwale.torque.values[i].split('@')[1]=='':
            carwale['torque_rpm'].iloc[i]=0
            carwale['torque'].iloc[i]=carwale.torque.values[i].split('@')[0].split()[0]
for i in range(len(carwale.price.values)):
    #print(carwale.price.values[i])
    if len(carwale.price.values[i].split())>2:
        if carwale.price.values[i].split()[2]=='Lakh':
            carwale['price'].iloc[i]=int(float(carwale.price.values[i].split()[1])*100000)
        elif carwale.price.values[i].split()[2]=='Crore':
            carwale['price'].iloc[i]=int(float(carwale.price.values[i].split()[1])*10000000)
for i in range(len(carwale.drive.values)):
    if carwale.drive.values[i]=='4WD / AWD':
        carwale['drive'].iloc[i]='AWD'
for i in range(len(carwale.fuel.values)):
    carwale['fuel'].iloc[i]=carwale.fuel.values[i].split()[0]
carwale
```
