```python
pagee=0
brand=[]
desp=[]
price=[]
dis=[]
while len(dis)<101:
    driver.get(page[pagee])
    brand_tag=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in brand_tag:
        brand.append(i.text)
    desp_tag=driver.find_elements_by_xpath("//div[@class='_2B099V']")
    for i in desp_tag:
        desp.append(i.text.split('\n')[1])
    price_tag=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in price_tag:
        price.append(i.text)
    dis_tag=driver.find_elements_by_xpath("//div[@class='_25b18c']")
    disc_tag=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    count=0
    for i in dis_tag[:40]:
        if i.text[-3:]=='off':
            dis.append(disc_tag[count].text)
            count+=1
        else:
            dis.append('-')
    pagee+=1
```

```python
url='https://www.myntra.com/shoes'
driver.get(url)
```

```python
search_btn=driver.find_elements_by_xpath("//li[@class ='colour-listItem']//label[@class ='common-customCheckbox']/div[@class = 'common-checkboxIndicator']")
search_btn[0].click()
```

```python
search_btn=driver.find_elements_by_xpath("//ul[@class ='price-list']//label[@class ='common-customCheckbox vertical-filters-label']/div[@class = 'common-checkboxIndicator']")
search_btn[1].click()
```

```python
brand=[]
desp=[]
price=[]
while True:
    brand_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h3[@class='product-brand']")
    for i in brand_tag:
        brand.append(i.text)
    desp_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//h4[@class='product-product']")
    for i in desp_tag:
        desp.append(i.text)
    price_tag=driver.find_elements_by_xpath("//div[@class= 'product-productMetaInfo']//div[@class='product-price']")
    for i in price_tag:
        price.append('Rs.'+i.text.split('Rs.')[1])
    if len(price)<100:
        driver.get(driver.find_element_by_xpath("//a[@rel = 'next']").get_attribute('href'))
    else:
        break
```

```python
dataframe=pd.DataFrame()
dataframe['Brand NAME']=brand[:100]
dataframe['Product Description']=desp[:100]
dataframe['Price']=price[:100]
dataframe
```
