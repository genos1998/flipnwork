```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import time
import calendar
from datetime import date
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
import ipyparallel as ipp
```

```python
month=1
year=2022
last_date=calendar.monthrange(year,month)[1]
```

```python
city=['GOI','PNQ','JAI']
destination=['BLR','BOM','DEL','HYD','AMD','CCU','AGR','MAA','GOI']
```

```python
names=[]
depart_times=[]
departs=[]
arr_times=[]
arrs=[]
durations=[]
stop=[]
prices=[]
datess=[]
monthss=[]
start_date='01'
mid_date='15'
mid_date_2='25'
datee=start_date
while len(names)<10000:
    print(datee,month)
    for home in city:
        for dest in destination:
            if home==dest:
                continue
            else:
                driver=webdriver.Chrome('chromedriver.exe')
                driver.maximize_window()
                url='https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination='+dest+'&destinationCountry=IN&flexi=0&flight_depart_date='+str(datee)+'%2F0'+str(month)+'%2F'+str(year)+'&hb=0&noOfSegments=1&origin='+home+'&originCountry=IN&type=O&unique=207953473702&viewName=normal'
                driver.get(url)
                time.sleep(33)
                for i in range(10):
                    driver.execute_script("var scrollingElement = (document.scrollingElement || document.body);scrollingElement.scrollTop = scrollingElement.scrollHeight;")
                    time.sleep(1)
                name=driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/span')
                depart_time=driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/div/div')
                depart=driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
                arr_time=driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[@class="bold fs-15 mb-2 pr time"]')
                arr=driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
                duration=driver.find_elements_by_xpath('//div[@class="stop-det bdr-btm pb-5 v-aligm-t text-left"]/div/p')
                stops=driver.find_elements_by_xpath('//div[@class="stop-det bdr-btm pb-5 v-aligm-t text-left"]/div/div/span')
                price=driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
                for i in range(len(name)):
                    names.append(name[i].text)
                    depart_times.append(depart_time[i].text)
                    departs.append(depart[i].text)
                    arr_times.append(arr_time[i].text)
                    arrs.append(arr[i].text)
                    durations.append(duration[i].text)
                    stop.append(stops[i].text)
                    prices.append(price[i].text)
                    datess.append(datee)
                    monthss.append(month)
            driver.close()
            print(len(names),end=' ')
    print()
    print(datee,month,len(names))
    print()
    if str(datee)==start_date:
        datee=mid_date
    elif str(datee)==mid_date:
        datee=mid_date_2
    elif str(datee)==mid_date_2:
        datee=last_date
    elif str(datee)==str(last_date):
        datee=start_date
        month+=1
        last_date=calendar.monthrange(year,month)[1]
```

```python
flight=pd.DataFrame({})
flight['Date']=datess
flight['Month']=monthss
flight['Name']=names
flight['Depart Times']=depart_times
flight['Departure']=departs
flight['Arrival Times']=arr_times
flight['Arrival']=arrs
flight['Duration']=durations
flight['No. of Stops']=stop
flight['price']=prices
flight
```

```python
for i in range(len(flight.Name)):
    flight['Arrival Times'].iloc[i]=flight['Arrival Times'].iloc[i].split()[0]
flight
```
